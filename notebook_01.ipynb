# Caminante aleatorio

Consiste en lanzar una moneda a cada paso y tomar una decisión hacía a donde caminar.
Lo que se hace es echar un volado y caminar hacia arriba o hacia abajo.

En este notebook se presenta

```python
from pylab import*
%matplotlib inline
```

```python
tam=1000; trayectoria=zeros(tam)
equis=range(tam); cero=array([0]*tam); pos=0
for i in range(tam):
    volado = rand()
    if volado<.5: pos=pos-1
    else: pos=pos+1
    trayectoria[i]=pos
plot(equis,trayectoria,equis,cero); show()
```

```python
x=linspace(.1,20)
y=cos(3*x)/x
y2=1./x
plot(x,y,x,y2,x,-y2)
show()
```

$|\frac{\cos{x}}{x}| \leq \frac{1}{x}$
