# Movimiento planetario

El movi planetario esta dado por la ley d gravitacion universal

$$m_p \frac{d^2\vec r}{dt^2} = -G\ M_s\ m_p \frac{\vec r}{|\vec r|^3} $$

El movi no depende de la masa del planeta, $m_p$, entonces podemos hacer el cambio de variable:

$T= \sqrt{\frac {G\ M_s}{Rt^3}} t$, y la ecuación anterior nos queda como

$$\frac{d^2\vec R}{dT^2} = - \frac{\vec R}{|\vec R|^3} $$

donde los valores númericos de las constantes es:

$G = 6.674 \times 10^{-11} \tfrac{N\ m^2}{kg^2}$

$M_s = 1.9891 \times 10^{30}\ kg$

$R_T = 1.495 \times 10^{11}\ m$

```python
from pylab import*
from scipy.integrate import*
%matplotlib inline
```

```python
def gravitacion(r_v,t):
    # r_v = [x, y, vx, vy]
    r = sqrt(r_v[0]**2+r_v[1]**2)
    dvx = -r_v[0]/r**3
    dvy = -r_v[1]/r**3
    dx = r_v[2]
    dy = r_v[3]
    return [dx, dy, dvx, dvy]
```

```python
time = linspace(0,2*pi-.1)
sol = odeint(gravitacion, [1,0, 0,1], time)
x, y = sol[:,0], sol[:,1]
axes().set_aspect('equal') 
xlim([-1.1,1.1]), ylim([-1.1,1.1])
plot(x,y,'or', 0,0,'oy')
show()
```
